Работа с пропущенными значениями

```python
# для начала проверим, есть ли в датасете пропущенные значения
data.isna().sum()
```

Пропущенных значений практически нет, сгенерируем их.

```python
import random
import numpy as np

columns = data.columns
for i in range(100):
    data.loc[random.randint(0, data.shape[0] -1), columns[random.randint(0, data.shape[1] -1)]] = np.nan
```

```python
data.isna().sum()
```

Видим, что появились пропуски. Теперь удалим строки с пропусками в колонках 'genre_names', 'riginal_title', 'id', 'release_date', 'overview'

```python
data = data.dropna(subset=["id", "original_title", "genre_names", "release_date", "overview"])
```

```python
data.isna().sum()
```

```python
data.shape
```

В числовых колонках заполним пропуски средним значением по столбцу

```python
data[columns[4]] = data[columns[4]].fillna(data[columns[4]].mean())
data[columns[5]] = data[columns[5]].fillna(data[columns[5]].mean())
data[columns[6]] = data[columns[6]].fillna(data[columns[6]].mean().round(0))
```

```python
data.isna().sum()
```

Создание нового поля, вычисленного на основе значений других полей

```python
#через выражение на базе имеющихся колонок
# success = popularity * vote_average
data["success"] = (data["popularity"] * data["vote_average"]).round(3)
data.info()
```

```python
# через DataFrame.apply
data["vote_average, %"] = data.apply(lambda row: row["vote_average"] * 10, axis=1)
data.info()
```

```python
# через DataFrame.apply
data["is_new"] = data["release_date"].apply(lambda e:  e > '2023-01-01')
data.info()
```

```python
bins = [0, 4, 7, 10]
labels = ['low', 'medium', 'high']
data['rating'] = pd.cut(data['vote_average'], bins=bins, labels=labels)
```

**№5.** Решейпинг данных с использованием .pivot (можно подать на вход результаты агрегатов, полученных ранее с использованием .groupby (сгруппировать по двум полям).

```python
# группировка по признау "новизны" фильма и по рейтингу. подсчет средней "порулярности" по группе
grouped = data.groupby(['is_new', 'rating'], as_index=False)['popularity'].mean()
grouped
```

```python
grouped.pivot(columns='rating', index='is_new', values='popularity')
```

**№6.** Решейпинг данных, соединённых с группировкой / агрегацией (одним словом - сводная таблица): .pivot_table. Группировать только по категориальным полям или числовым, если значений немного.

```python
# загрубим значение даты выхода для группировке по годам
data['release_year'] = data["release_date"].apply(lambda e:  str(round(int(e[0:4]), -1)) + '-e')
```

```python
data.pivot_table(values='popularity', index='release_year', columns='rating', aggfunc='mean')
```

**№11.** Продемонстировать работу `.merge`

```python
#Создадим фрейм с данными о популярности каждого жанра
genge_popularity = data['main_genre'].unique()
popularity = np.random.randint(1, 11, genge_popularity.shape)
genge_popularity_frame = pd.DataFrame(
    {
        'main_genre': genge_popularity,
        'genge_popularity': popularity,
    }
)
genge_popularity_frame
```
